````python
%matplotlib inline
from matplotlib import pyplot as plt
from PIL import Image
import glob
import math
````

````python
img_list = sorted(glob.glob('*'), reverse=False)
img_1334 = img_list[:10]
img_1370 = img_list[10:20]
img_2886 = img_list[20:30]
img_816 = img_list[30:]
````

````python
plt.figure(figsize=(18,6))
plt.suptitle('a lighted kitchen looks shiny clean and nice.')
for i in range(len(img_1334)):
    plt.subplot(2, 5, i+1)
    img = Image.open("{}".format(img_1334[i]))
    plt.imshow(img)
plt.show()
````

````python
plt.figure(figsize=(18,6))
plt.suptitle('a bus turning a corner on a city street.')
for i in range(len(img_1370)):
    plt.subplot(2, 5, i+1)
    img = Image.open("{}".format(img_1370[i]))
    plt.imshow(img)
plt.show()
````
